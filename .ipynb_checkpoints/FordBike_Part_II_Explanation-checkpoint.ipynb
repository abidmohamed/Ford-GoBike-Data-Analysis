```python
filepath = "201902-fordgobike-tripdata_clean.csv"
df = pd.read_csv(filepath)
```

```python
df.info()
```

## Does gender affect the duration of the ride in the weekdays ?

> The next Graphs shows the relation between rides, times and gender.


```python
# selecting columns to plot 
cols=['start_day_of_week', 'end_day_of_week', 'duration_sec', 'member_gender']
# creating new data
plot_data = df[cols]
# Melting data
plot_data = plot_data.melt(id_vars=['duration_sec', 'member_gender'], var_name='start/end', value_name='day')
plot_data.head(2)
```

```python
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
```

```python
## Plotting
g=sns.catplot(
    data=plot_data, 
    x="day", 
    y="duration_sec", 
    hue="member_gender",
    col="start/end",
    order=day_order,
    kind="bar",
) 
g.set_xticklabels(rotation=30)
g.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
g.fig.suptitle('Ride duration distrbution based on week days and gender', fontsize=20)
g.fig.set_size_inches(20,10);
```

## What is the relation between user type and ride duration and days ? 

> Does The day matter for having a ride ? the next graphs shows the relation between the day and the ride

```python
# selecting columns to plot 
cols=['start_day_of_week', 'end_day_of_week', 'duration_sec', 'user_type']
# creating new data
plot_data = df[cols]
# Melting data
plot_data = plot_data.melt(id_vars=['duration_sec', 'user_type'], var_name='start/end', value_name='day')
plot_data.head(2)
```

```python
## Plotting
g=sns.catplot(
    data=plot_data, 
    x="day", 
    y="duration_sec", 
    hue="user_type",
    col="start/end",
    order=day_order,
    kind="bar",
);
g.set_xticklabels(rotation=30)
g.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
g.fig.suptitle('Ride duration distrbution based on week days and user type', fontsize=20)
g.fig.set_size_inches(20,10);
```

## How time, gender and user type are related ?

The next graphs shows the relations between gender and user type and the ride time


```python
# Start hour
g=sns.relplot(
    data=df, kind="line",
    x="start_hour", y="duration_sec", col="member_gender",
    hue="user_type", style="user_type", markers=True,
)
g.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
g.fig.suptitle('Ride duration distrbution based on start hour, user type and gender', fontsize=20)
g.fig.set_size_inches(20,10);
```

```python
# End hour
g=sns.relplot(
    data=df, kind="line",
    x="end_hour", y="duration_sec", col="member_gender",
    hue="user_type", style="user_type", markers=True,
)
g.fig.subplots_adjust(top=0.9) # adjust the Figure in rp
g.fig.suptitle('Ride duration distrbution based on end hour, user type and gender', fontsize=20)
g.fig.set_size_inches(20,10);
```
